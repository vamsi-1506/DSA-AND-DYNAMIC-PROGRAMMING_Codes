```python
class stack:
    def __init__(self):
        self.l=[]
        self.size=5
        self.top=-1
        
    def push(self,x):
       if self.top<self.size-1: 
        self.l.append(x)
        self.top+=1
       else:
         print("overflow")
           
    def display(self):
       print(self.l)
    def pop(self):
       if self.top!=-1: 
        self.l.pop()
        self.top-=1
       else:
           print("u")
           
    def peek(self):
        if self.top!=-1: 
          print(self.l[self.top])
        else:
            print("no")

c=stack()
c.push(9)
c.push(78)
c.push(78)
c.push(78)
c.push(78)
c.push(78)
c.pop()
c.pop()
c.pop()
c.pop()
c.pop()
c.pop()
c.display()
c.peek()
```

```python
class queue:
    def __init__(self,size=3):
        self.l=[]
        self.size=3
        self.front=self.rear=-1
    def enqueue(self,data):
       if self.rear<self.size-1:
        self.l.append(data)
        self.rear+=1
        if self.front==-1:
            self.front+=1
       else:
           print("queue is full")
    def dequeue(self):
      if self.front<=self.rear and self.front!=-1:
          data=self.l[self.front]
          self.l[self.front]=None 
          self.front+=1
          if self.front>self.rear:
              self.front=self.rear=-1
          return data #(optional)
      else:
          print("queue is empty")
    def peek(self):
        if self.front<=self.rear and self.rear!=-1:
              print(self.l[self.front])
        else:
            print("queue is empty")
    def display(self):
      if self.front<self.rear:
        for i in range(self.front,self.rear+1):
            print(self.l[i])
      else:
          print("queue is empty")
c=queue()
c.enqueue(1)
c.enqueue(2)
c.enqueue(3)
c.dequeue()
c.peek()
c.display()
```

```python
#Binary search tree
class node:
    def __init__(self,key):
        self.left=None
        self.key=key
        self.right=None

class Binarysearch:
    def __init__(self):
        self.root=None
        
    def insert(self,key):
        self.root=self.__insert_recursive(self.root,key)   #private function
        
    def __insert_recursive(self,root,key):   #(private function/protector)
        if root is None:
            return node(key)
        if key<root.key:
            root.left=self.__insert_recursive(root.left,key)
        elif key>root.key:
            root.right=self.__insert_recursive(root.right,key)
        return root

    def inorder(self,root):
       if root:
          self.inorder(root.left)
          print(root.key,end=" ")
          self.inorder(root.right)
          
   # inorder-l,ro,ri
   # postorder-l,ri,ro
   # preorder-ro,l,ri

    def __min_value_node(self, root):
        current = root
        while current.left is not None:
            current = current.left
        return current

    def search_recursive(self,root,val):
        if not root or root.key==val:
          if root is None:
              print('not found')
              return
          else:
              print('found')
              return
        if val<root.key:
            return  self.search_recursive(root.left,val)
        else:
            return  self.search_recursive(root.right,val)

    def height(self,root):
        if root is None:
            return 0
        lh=self.height(root.left)
        lr=self.height(root.right)
        return 1+max(lh,lr)
                
    def level_max(self,root):
        temp=[]
        temp.append(root)
        res=[]
        while len(temp)>0:
            maxi=0
            for i in range(len(temp)):
                if temp[0].left is not None:
                    temp.append(temp[0].left)
                if temp[0].right is not None:
                    temp.append(temp[0].right)
                maxi=max(temp[0].key,maxi)
                temp.pop(0)
            res.append(maxi)
        print('max at each level:',res)

    def delete(self, root, key):
        if root is None:
            return root
        if key < root.key:
            root.left = self.delete(root.left, key)
        elif key > root.key:
            root.right = self.delete(root.right, key)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            temp = self.__min_value_node(root.right)
            root.key = temp.key
            root.right = self.delete(root.right, temp.key)
        return root
        
bs=Binarysearch()
bs.insert(3)
bs.insert(2)
bs.insert(1)
bs.insert(4)
bs.insert(5)
bs.inorder(bs.root)
#print(bs.__min_value_node(bs.root))
bs.search_recursive(bs.root,3)
print('height:',bs.height(bs.root))
bs.level_max(bs.root)
bs.delete(bs.root,1)
bs.inorder(bs.root)
```

```python
#directed graph
class dgraph:
    def transitiveclosure(self,s): #(Floyd-Warshall algorithm)
        a=s.copy()
        v=len(a)
        for k in range(v):
            for i in range(v):
                for j in range(v):
                    a[i][j]=a[i][j] or a[i][k] and a[k][j]
        print(a)
            
dg=dgraph()
s=[[1,1,0,1],
   [0,1,1,1],
   [0,0,1,0],
   [0,0,0,1]]
dg.addedges(s)
```

```python
def bubblesort(a):
    for i in range(len(a)-1):
        for j in range(len(a)-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    print(a)
bubblesort([3,4,2,5,1])
```

```python
def selectionsort(a):
    for i in range(len(a)-1):
        min_index=i
        for j in range(i+1,len(a)):
            if a[j]<a[min_index]:
                min_index=j
        a[i],a[min_index]=a[min_index],a[i]
    print(a)
selectionsort([3,4,2,5,1])
```
